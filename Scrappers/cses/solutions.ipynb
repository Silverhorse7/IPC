```python
import os 
import csv
import re
import pandas as pd 
```

```python
df = pd.DataFrame(columns=['source', 'problem_link', 'solution'])
root = "https://cses.fi/problemset/task/"
```

```python
with open('name_id_mapping.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader, None)  # Skip the header
    name_id_map = {rows[0]: rows[1] for rows in reader}  # Create a dictionary {Name: PID}
```

```python
source = "https://github.com/Jonathan-Uy/CSES-Solutions"
repo_path = 'repos/Jonathan-Uy'

for root_dir, dir_name, files in os.walk(repo_path):
    for file in files:
        if file[-4:] != ".cpp":
            continue

        file_path = os.path.join(root_dir, file)
        solution = open(file_path).read()

        pid = name_id_map.get(file[:-4])

        if pid == None:
            continue

        problem_link = root + str(pid)

        new_row = {
            'source': source,
            'problem_link': problem_link,
            'solution': solution
        }

        df.loc[len(df)] = new_row
```

```python
import pandas as pd
```

```python
cses_sol = pd.read_csv('datasets/cses_solutions.csv')
leetcode_sol = pd.read_csv('datasets/leetcode_solutions.csv')
uva_sol = pd.read_csv('datasets/uva_solutions.csv').rename(columns={'Solution': 'solution', 'Problem Link': 'problem_link', 'Repository': 'source'})
yosupo_sol = pd.read_csv('datasets/yosupo_solutions.csv')
```

```python
cses_problem = pd.read_csv('datasets/cses_problems.csv')
leetcode_problem = pd.read_csv('datasets/leetcode_problems.csv')
uva_problem = pd.read_csv('datasets/uva_problems.csv')
yosupo_problem = pd.read_csv('datasets/yosupo_problems.csv')
```

```python
cses_sol.head()
```

```python
cses_sol['problem_link'] =[i.replace('https://cses.fi', '') for i in cses_sol['problem_link']]
#print(cses_sol['problem_link'])

cses = pd.merge(cses_problem, cses_sol, on='problem_link').drop_duplicates(subset='problem_link', keep='first')
cses.head()
```

```python
cses.info()
```

```python
leetcode_sol['problem_link'] =[i.replace('https://leetcode.com/problems/', '') for i in leetcode_sol['problem_link']]
print(leetcode_sol['problem_link'])

leetcode = pd.merge(leetcode_problem, leetcode_sol, left_on='titleSlug',right_on='problem_link').drop_duplicates(subset='problem_link', keep='first')
leetcode.head()
```

```python
leetcode.info()
```

```python
uva = pd.merge(uva_problem, uva_sol, on='problem_link').drop_duplicates(subset='problem_link', keep='first')
uva.head()
```

```python
uva.info()
```

```python
yosupo = pd.merge(yosupo_problem, yosupo_sol, on='problem_link').drop_duplicates(subset='problem_link', keep='first')
yosupo.head()
```

```python
yosupo.info()
```

```python
cses = cses[['problem_statement', 'solution']].rename(columns={'problem_statement': 'problem'})
leetcode = leetcode[['problem_text', 'solution']].rename(columns={'problem_text': 'problem'})
uva = uva[['problem_text', 'solution']].rename(columns={'problem_text': 'problem'})
yosupo = yosupo[['problem_statement', 'solution']].rename(columns={'problem_statement': 'problem'})
```

```python
cses.head()
```

```python
combined_data = pd.concat([cses, leetcode, uva, yosupo], axis=0)
```

```python
combined_data.info()
```
